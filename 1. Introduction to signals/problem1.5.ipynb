Consider the following analog sinusoidal signal: \
$x_{a}(t) = 3 \cdot sin(100\pi t) $


a) Sketch the signal $x_{a}(t)$ for  $0 \le t \le 30 ms$. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
START_TIME = 0
END_TIME = 30e-3 # 30ms

t = np.linspace(START_TIME, END_TIME, 1000)
xa = 3 * np.sin(100 * np.pi *t)
```

```python
plt.figure()
plt.plot(t*1000, xa)
plt.xlim(START_TIME*1000, END_TIME*1000)
plt.title("Xa(t)")
plt.xlabel("time (ms)")
plt.ylabel("amplitude")
```
