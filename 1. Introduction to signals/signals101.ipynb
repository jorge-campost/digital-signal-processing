# Ejercicio 1:
Considere la señal sinusoidal analógica:


$x_{a}(t) = 5 \cdot cos(400\pi t +  \cfrac{\pi}{6})$. Considerando un  tiempo para 0< t < 20ms. 

Se pide:
1. La frecuencia de la señal analógica en Hz.
2. Si la señal se muestrea a la frecuencia Fs = 2000 muestras/s determine la frecuencia de la señal discreta f.
3. Duración de la señal en segundos Tf.
4. Cantidad de muestras N.
5. Vector de muestras n.
6. Dibuje la señal muestreada como una función de muestras, use el comando que permita visualizar las muestras.
7. Dibuje la señal muestreada como una función del tiempo.
8. ¿Cuantos periodos se grafican?
9. ¿Cuantas muestras por periodo se tienen?

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

1. La frecuencia de la señal analógica en Hz

```python
F = 200  # 200 Hz
```

2. Si la señal se muestrea a la frecuencia Fs = 2000 muestras/s. Determine frecuencia de la señal discreta f

```python
Fs = 2000
f = F / Fs
```

3. Duración de la señal analógica en segundos Tf

```python
Tf = 0.02  # 20 ms
```

4. Cantidad de muestras N

```python
N = Tf * Fs
print(f"La cantidad de muestras es {N}")
```

5. Vector de muestras

```python
n = np.arange(N, dtype=int)
print(n)
```

6. Dibuje la señal muestreada como una función de muestras. Use el comando que permita visualizar las muestras.

```python
x = 5 * np.cos(2 * math.pi * F * (n / Fs) + math.pi / 6)
```

7. Dibuje la señal muestreada como una función de tiempo

```python
plt.figure()
plt.stem(n, x)
plt.xlabel("n° de muestras")
plt.ylabel("amplitud")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.title("x(n)")
plt.show()
```

```python
time_ms = (n / Fs) * 1000
plt.figure()
plt.stem(time_ms, x)
plt.xlabel("time (ms)")
plt.ylabel("amplitude")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.title("x(n)")
plt.show()
```

8. ¿Cuántos periodos se grafican? 4
9. ¿Cuántas muestras por periodo se tiene? 10

10. Comparar la señal discreta con una 'analógica'


```python
t = np.linspace(0, 0.02, 1000)
xa = 5 * np.cos(400 * math.pi * t + math.pi / 6)

plt.figure()
plt.plot(t, xa, linestyle="--", color="red")
plt.stem(n / Fs, x)
plt.xlabel("time (s)")
plt.ylabel("amplitude")
plt.title("x(n) vs x(t)")
plt.show()
print(t.shape)
```
