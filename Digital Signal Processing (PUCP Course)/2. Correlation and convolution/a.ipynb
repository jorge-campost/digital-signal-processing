# Convolution


```python
import numpy as np
import matplotlib.pyplot as plt
```

Representing signals $x(n)$ y $h(n)$ as arrays of pairs (n,x(n))


```python
x = np.array([[0, 1], [1, 2], [2, 3], [3, 1]])

h = np.array([[-1, 1], [0, 2], [1, 1], [2, -1]])

fig = plt.figure()
plt.plot(x[:, 0], x[:, 1], label="x(n)", marker="o")
plt.plot(h[:, 0], h[:, 1], label="h(n)", marker="o")
plt.axhline(0, color="black", linestyle="--")
plt.axvline(0, color="black", linestyle="--")
plt.xlabel("n")
plt.ylabel("amplitude")
plt.legend()
plt.show()
```

```python
y = np.convolve(x[:, 1], h[:, 1])
fig = plt.figure()
plt.plot(y, marker="o")
plt.axhline(0, color="black", linestyle="--")
plt.axvline(0, color="black", linestyle="--")
plt.show()
```

The vector "y" (the convolution between "h" and "x") assumes both inputs start at zero.


```python
y_length = len(y)
print(y_length)
```

Let's add son zeros in order to make both signals start at the same "x"(number of sample)


```python
x = np.array([[-1, 0], [0, 1], [1, 2], [2, 3], [3, 1]])
h = np.array([[-1, 1], [0, 2], [1, 1], [2, -1], [3, 0]])

fig = plt.figure()
plt.plot(x[:, 0], x[:, 1], label="x(n)", marker="o")
plt.plot(h[:, 0], h[:, 1], label="h(n)", marker="o")
plt.axhline(0, color="black", linestyle="--")
plt.axvline(0, color="black", linestyle="--")
plt.xlabel("n")
plt.ylabel("amplitude")
plt.legend()
plt.show()
```

It also neccesary to create a vector n.


```python
y = np.convolve(x[:, 1], h[:, 1])

y_length = len(y)
n = np.arange(-2, y_length - 2)

fig = plt.figure()
plt.plot(n, y, marker="o")
plt.axhline(0, color="black", linestyle="--")
plt.axvline(0, color="black", linestyle="--")
plt.xlabel("n")
plt.ylabel("amplitude")
plt.title("y(n)")
plt.show()
```
